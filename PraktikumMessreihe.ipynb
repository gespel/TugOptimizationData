```python
import matplotlib.pyplot as plt
import numpy as np

# normale Messreihe: 1632, 4019, 986
messreihe_1 = {
    "BARITE_200": [1632, 1630, 1590],
    "BARITE_LARGE": [4019, 4040, 4032],
    "SURFEX": [986, 920, 935]
}

messreihe_2 = {
    "BARITE_200": [1650, 1700, 1610],
    "BARITE_LARGE": [4200, 4140, 4137],
    "SURFEX": [1213, 1205, 1196]
}

def plot_comparison(messreihe_1, messreihe_2):
    labels = ["BARITE_200", "BARITE_LARGE", "SURFEX"]

    messwerte_1 = [np.mean(messreihe_1[label]) for label in labels]
    messwerte_2 = [np.mean(messreihe_2[label]) for label in labels]

    x = np.arange(len(labels))  # die Labelpositionen
    width = 0.35  # die Breite der Balken

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, messwerte_1, width, label='Eigen::SparseMatrix')
    rects2 = ax.bar(x + width/2, messwerte_2, width, label='std::vector')

    # Diagrammanpassungen
    ax.set_ylabel('Laufzeit in ms')
    ax.set_title('Eigen::SparseMatrix vs std::vector')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    plt.savefig("spm_vs_stdvec.jpg")
    plt.show()
    

plot_comparison(messreihe_1, messreihe_2)
```
